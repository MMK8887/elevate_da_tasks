```python
import pandas as pd

train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')

display(train_df.head())
display(test_df.head())
```

```python
print("--- Train DataFrame ---")
display(train_df.describe())
display(train_df.info())

print("\n--- Value Counts for Train DataFrame ---")
for col in ['Survived', 'Pclass', 'Sex', 'Embarked']:
    if col in train_df.columns:
        print(f"\nValue counts for {col}:")
        display(train_df[col].value_counts())

print("\n--- Test DataFrame ---")
display(test_df.describe())
display(test_df.info())

print("\n--- Value Counts for Test DataFrame ---")
for col in ['Pclass', 'Sex', 'Embarked']:
    if col in test_df.columns:
        print(f"\nValue counts for {col}:")
        display(test_df[col].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Histograms for numerical columns
numerical_cols = ['Age', 'Fare', 'SibSp', 'Parch']
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 2, i + 1)
    sns.histplot(train_df[col].dropna(), kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

# 2. Box plots for numerical columns against categorical columns
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
sns.boxplot(x='Survived', y='Age', data=train_df)
plt.title('Age vs Survived')

plt.subplot(2, 2, 2)
sns.boxplot(x='Pclass', y='Fare', data=train_df)
plt.title('Fare vs Pclass')

plt.subplot(2, 2, 3)
sns.boxplot(x='Survived', y='Fare', data=train_df)
plt.title('Fare vs Survived')

plt.subplot(2, 2, 4)
sns.boxplot(x='Pclass', y='Age', data=train_df)
plt.title('Age vs Pclass')

plt.tight_layout()
plt.show()


# 3. Scatter plots for pairs of numerical columns
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Age', y='Fare', data=train_df)
plt.title('Age vs Fare')
plt.show()


# 4. Pair plot for a subset of relevant numerical columns
pairplot_cols = ['Survived', 'Pclass', 'Age', 'Fare']
sns.pairplot(train_df[pairplot_cols].dropna())
plt.suptitle('Pairwise Relationships and Distributions', y=1.02)
plt.show()


# 5. Correlation matrix heatmap
correlation_matrix = train_df[numerical_cols + ['Survived', 'Pclass']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Columns')
plt.show()
```

```python
# 1. Fill missing 'Age' values
train_df['Age'].fillna(train_df['Age'].median(), inplace=True)
test_df['Age'].fillna(train_df['Age'].median(), inplace=True) # Use median from train_df for consistency

# 2. Fill missing 'Embarked' values in train_df
most_frequent_embarked = train_df['Embarked'].mode()[0]
train_df['Embarked'].fillna(most_frequent_embarked, inplace=True)

# 3. Fill missing 'Fare' values in test_df
test_df['Fare'].fillna(train_df['Fare'].median(), inplace=True) # Use median from train_df

# 4. Drop the 'Cabin' column
train_df.drop('Cabin', axis=1, inplace=True)
test_df.drop('Cabin', axis=1, inplace=True)

# Verify the changes
print("Missing values after filling and dropping:")
print("Train DataFrame:")
display(train_df.isnull().sum())
print("\nTest DataFrame:")
display(test_df.isnull().sum())
```

```python
# 1. Create 'FamilySize'
train_df['FamilySize'] = train_df['SibSp'] + train_df['Parch'] + 1
test_df['FamilySize'] = test_df['SibSp'] + test_df['Parch'] + 1

# 2. Create 'IsAlone'
train_df['IsAlone'] = (train_df['FamilySize'] == 1).astype(int)
test_df['IsAlone'] = (test_df['FamilySize'] == 1).astype(int)

# 3. Create categorical features for 'Age' by binning
# Define age bins
age_bins = [0, 12, 18, 35, 60, 80]
age_labels = ['Child', 'Teenager', 'Young Adult', 'Adult', 'Senior']
train_df['AgeBin'] = pd.cut(train_df['Age'], bins=age_bins, labels=age_labels, right=False)
test_df['AgeBin'] = pd.cut(test_df['Age'], bins=age_bins, labels=age_labels, right=False)

# 4. Create categorical features for 'Fare' by binning
# Use qcut to create quantile-based bins for Fare
# Handle potential errors if bins have identical values (e.g., many zero fares)
try:
    train_df['FareBin'] = pd.qcut(train_df['Fare'], q=4, labels=['Low', 'Medium', 'High', 'Very High'], duplicates='drop')
    test_df['FareBin'] = pd.qcut(test_df['Fare'], q=4, labels=['Low', 'Medium', 'High', 'Very High'], duplicates='drop')
except ValueError as e:
    print(f"Could not create 4 Fare bins due to data distribution: {e}")
    print("Attempting to create fewer bins or using fixed width bins.")
    # Fallback to fewer bins if 4 is not possible
    try:
        train_df['FareBin'] = pd.qcut(train_df['Fare'], q=3, labels=['Low', 'Medium', 'High'], duplicates='drop')
        test_df['FareBin'] = pd.qcut(test_df['Fare'], q=3, labels=['Low', 'Medium', 'High'], duplicates='drop')
    except ValueError as e_fallback:
        print(f"Could not create 3 Fare bins either: {e_fallback}")
        # Fallback to fixed width bins if quantile binning fails
        fare_bins_fixed = [0, 10, 30, 100, 600]
        fare_labels_fixed = ['Very Low', 'Low', 'Medium', 'High']
        train_df['FareBin'] = pd.cut(train_df['Fare'], bins=fare_bins_fixed, labels=fare_labels_fixed, right=False, include_lowest=True)
        test_df['FareBin'] = pd.cut(test_df['Fare'], bins=fare_bins_fixed, labels=fare_labels_fixed, right=False, include_lowest=True)
        print("Used fixed width bins for Fare.")


# 5. Display the first few rows and data types
print("Train DataFrame with new features:")
display(train_df.head())
print("\nTrain DataFrame Info:")
display(train_df.info())

print("\nTest DataFrame with new features:")
display(test_df.head())
print("\nTest DataFrame Info:")
display(test_df.info())
```


## Summary:

### Data Analysis Key Findings

*   The training dataset contains 891 entries, while the test dataset contains 418 entries. Both datasets have a mix of data types (integer, float, object).
*   Missing values are present in the 'Age' and 'Cabin' columns in both datasets. 'Embarked' in the train set and 'Fare' in the test set also have missing values.
*   Value counts show that the majority of passengers in the training set did not survive, most passengers were in Pclass 3, and there were significantly more male than female passengers. The most common embarkation point was 'S'.
*   The distribution of 'Fare' is highly skewed to the right, with most passengers paying lower fares.
*   Passengers who survived tend to have paid higher fares, and Pclass 1 is associated with higher fares.
*   Pclass and Fare appear to be key features influencing survival: Pclass has a negative correlation with Survived (-0.34), and Fare has a positive correlation with Survived (0.26).
*   Age has a weak negative correlation with Survived (-0.08).

*   Missing values in 'Age', 'Embarked', and 'Fare' were successfully filled using the median (Age, Fare) and mode (Embarked) from the training data. The 'Cabin' column was dropped from both dataframes.
*   New features 'FamilySize' (SibSp + Parch + 1) and 'IsAlone' (derived from FamilySize) were created.
*   'Age' was binned into five categories: 'Child', 'Teenager', 'Young Adult', 'Adult', and 'Senior'.
*   'Fare' was binned into categories ('Low', 'Medium', 'High', 'Very High' or 'Low', 'Medium', 'High' or fixed width bins) using quantile-based binning with a fallback mechanism.
*   Categorical features ('Sex', 'Embarked', 'AgeBin', 'FareBin') were one-hot encoded.
